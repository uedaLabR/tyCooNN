```python
train = 0; test = 0;
cnt = 18000
lim = 6000
unit = 1000
p = int(lim / unit)
q = int(p * 0.1667)
print(p,q)
for i in range(cnt):
    if i % p >= q:
        train += 1
    else:
        test += 1
    if i >= lim-1:
        break
train,test
```

```python
test / (test + train)
```
