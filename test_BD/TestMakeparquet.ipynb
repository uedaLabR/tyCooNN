Defining the location of trimmed pq file that are read from input basecalled reads.

```python
class genPQ:
    LIST_FILE = "/share/bhaskar/tyCooNNTest/inputs.txt"
    OUTPQ_DIR = "/share/bhaskar/tyCooNNTest/trimpq/"
    INP_DIR = "/share/trna/testdata/ecolibasecalled/"
    SETTINGS = "/share/bhaskar/tyCooNNTest/preprocess_setting.yaml"
    
    def __init__(self,settings_file = SETTINGS):
        self.settings = settings_file
        pass
    
    def writeIOFile(self,out_file=LIST_FILE,outpq_dir=OUTPQ_DIR,inp_dir=INP_DIR):

        fw = open(out_file,'w')
        
        self.out_file = out_file
        self.outpq_dir = outpq_dir
        self.inp_dir = inp_dir
        
        files = glob.glob(inp_dir+"/*/")
    
        for f in files:
            dirlist = []
            fs = glob.glob(f+"/*/*/*/*/workspace")
            if len(fs)>0:
                dirlist = fs
            else:
                fs = glob.glob(f + "/*/*/*/workspace")
                dirlist = fs
            basename = os.path.basename(os.path.dirname(f)).replace("ecoli_rcc_","")
            basename = basename.split("_")[1]
            pqpath = outpq_dir + basename + ".pq"
            lst = ",".join(dirlist)

            fw.write(basename + "\t" + lst + "\t" + pqpath + "\n")
        
        fw.close()
        
    def generatePq(self,takeCount=12000):
        print("Writting from: ",self.out_file)
        print("Use settings from: ",self.settings)
        return pqg.generatePqForTrainingAll(paramPath=self.settings, listOfIOPath=self.out_file, takeCount=takeCount)
        
```

```python
g = genPQ()
g.writeIOFile()
stat = g.generatePq(takeCount=18000)
```

```python
import pandas as pd
```

```python
trna = []
rows = []
for k in stat.keys():
    s = stat[k]
    type_data = []
    N = int(s['N'])
    for k1 in s.keys():
        if k1 != 'N':
            s[k1] /= N
    this_row = pd.DataFrame([s])
    rows.append(this_row)
    trna.append(k)
```

```python
dat = pd.concat(rows)
dat['name'] = trna
```

```python
dat = dat.sort_values(['pass'],ascending=False)
```
